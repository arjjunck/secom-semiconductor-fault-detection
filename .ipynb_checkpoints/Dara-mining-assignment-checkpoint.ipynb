<h2>3. Outlier Analysis</h2>

```python
#z-score , flagging values above standard deviation3 
z_scores = np.abs(stats.zscore(X_scaled))
outlier_mask_zscore = (z_scores > 3).any(axis=1)
print(f"Outliers detected (Z-score > 3): {outlier_mask_zscore.sum()} samples")
#for detecting global outliers
```

```python
#local outlier factor , density based
pca_lof = PCA(n_components=30)
X_lof = pca_lof.fit_transform(X_scaled)
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
lof_preds = lof.fit_predict(X_lof)  #-1 =outlier, 1=inlier
outlier_mask_lof = lof_preds == -1
print(f"Outliers detected (LOF): {outlier_mask_lof.sum()} samples")
#for detecting local outliers, how far a point is from outlier
```

```python
#outlier visualisation
plt.figure(figsize=(10, 6))
plt.scatter(X_lof[~outlier_mask_lof, 0], X_lof[~outlier_mask_lof, 1],c='steelblue', label='Inlier', alpha=0.5, s=20)
plt.scatter(X_lof[outlier_mask_lof, 0], X_lof[outlier_mask_lof, 1],c='tomato', label='Outlier (LOF)', alpha=0.8, s=40, marker='x')
plt.title('LOF Outlier Detection (PCA 2D projection)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.tight_layout()
plt.show()

# Decision: remove outliers to improve model generalisation, only local outlier is removed
keep_mask = ~outlier_mask_lof
X_clean = X_scaled[keep_mask]
y_clean = y.values[keep_mask]
print(f"Samples after outlier removal: {X_clean.shape[0]}")
```

<h1>4. Dimensionality Reduction (PCA)</h2>

```python
#1. Bartlett’s Test of Sphericity
#to checks if the variables are identity Matrix
chi_square_value, p_value = calculate_bartlett_sphericity(X_clean)
print(f"Chi-square: {chi_square_value:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: The variables are correlated enough for PCA.")
else:
    print("Result: The variables are not correlated enough for PCA.")
```

> chi square and p-value confirm the variables are perfectly correlated

```python
# 2. Kaiser-Meyer-Olkin(KMO) Test
kmo_per_variable, kmo_total = calculate_kmo(X_clean)
print(f"KMO Score: {kmo_total:.4f}")

if ((kmo_total > 0.6) | np.isnan(kmo_total)):
    print("Result: Dataset is good for PCA")
else:
    print("Result: Dataset is not good for PCA")
```

```python
#fit PCA
pca_full = PCA()
pca_full.fit(X_clean)

cum_variance = np.cumsum(pca_full.explained_variance_ratio_) #cumilative variance
n_95 = np.argmax(cum_variance >= 0.95) +1
print(f"Components to retain 95% variance:{n_95}")
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(pca_full.explained_variance_ratio_[:50], marker='o', markersize=4)
axes[0].set_title('Scree Plot')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Explained Variance Ratio')

axes[1].plot(cum_variance[:100])
axes[1].axhline(0.95, color='red', linestyle='--', label='95% threshold')
axes[1].axvline(n_95, color='green', linestyle='--', label=f'n={n_95}')
axes[1].set_title('Cumulative Explained Variance')
axes[1].set_xlabel('Number of Components')
axes[1].set_ylabel('Cumulative Variance')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
#reducing to 95% varance component
pca = PCA(n_components=n_95)
X_pca = pca.fit_transform(X_clean)
print(f"Reduced shape: {X_pca.shape}")
plt.figure(figsize=(9, 6))
for cls, label, colour in zip([0, 1],['Pass','Fail'],['steelblue','tomato']):
    mask = y_clean == cls
    plt.scatter(X_pca[mask, 0], X_pca[mask, 1],
                label=label, alpha=0.5, s=20, c=colour)
plt.title('PCA – First Two Principal Components')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.tight_layout()
plt.show()
```

<h2>5. Clustering </h2>

```python
#to find optimal number of clusters
inertia = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_pca)
    inertia.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.title('K-Means Elbow Plot')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia') #how spread out the clusters are
plt.xticks(K_range)
plt.tight_layout()
plt.show()
```

> lack of clear elbow in the k-means elbow plot, after 3 most points are flat, suggesting data may be represnted by 3 clusters

```python
# Fit K-Means with k=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
km_labels = kmeans.fit_predict(X_pca)

plt.figure(figsize=(9, 6))
# Changed cmap to 'viridis' or 'Set1' because 'coolwarm' is better for 2 classes
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=km_labels, cmap='viridis', alpha=0.5, s=20)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='black', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering (k=3) on PCA Space')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.tight_layout()
plt.show()

#comparison of all 3 clusters with true labels
table = pd.crosstab(km_labels, y_clean)

for i in table.index:
    print(f"--- Cluster {i+1} Summary ---")
    print(f"Pass (0): {table.loc[i, 0]}")
    print(f"Fail (1): {table.loc[i, 1]}")
    print(f"Total   : {table.loc[i].sum()}\n")
```

```python
#Hierarchical Clustering
np.random.seed(70) 
sample_idx = np.random.choice(len(X_pca), 200, replace=False)
X_sample = X_pca[sample_idx, :10]

linked = linkage(X_sample, method='ward')

plt.figure(figsize=(14, 6))
dendrogram(linked, truncate_mode='level', p=5, leaf_rotation=90, leaf_font_size=8)
plt.title('Hierarchical Clustering Dendrogram (Ward linkage, subsample n=200)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

> This supports earlier k means finding, confirming dataset partitioning into 3 groups naturally

<h2>6. Classification</h2>

```python
# Train/test split — stratified to preserve class ratio
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y_clean, test_size=0.2, random_state=42, stratify=y_clean
)
print(f"Train: {X_train.shape} | Test: {X_test.shape}")
print(f"Train class balance: {np.bincount(y_train)}")
print(f"Test  class balance: {np.bincount(y_test)}")
```

```python
# --- k-Nearest Neighbours ---
knn = KNeighborsClassifier(n_neighbors=7, metric='euclidean')
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

print("=== k-NN (k=7) ===")
print(classification_report(y_test, y_pred_knn, target_names=['Pass', 'Fail']))
```

```python
# --- Decision Tree ---
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

print("=== Decision Tree (max_depth=5) ===")
print(classification_report(y_test, y_pred_dt, target_names=['Pass', 'Fail']))
```

```python
# --- Support Vector Machine ---
svm = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)

print("=== SVM (RBF kernel) ===")
print(classification_report(y_test, y_pred_svm, target_names=['Pass', 'Fail']))
```

<h2>7. Model Evaluation</h2>

```python
# Confusion matrices side by side
models = {
    'k-NN': y_pred_knn,
    'Decision Tree': y_pred_dt,
    'SVM': y_pred_svm
}

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
for ax, (name, preds) in zip(axes, models.items()):
    cm = confusion_matrix(y_test, preds)
    disp = ConfusionMatrixDisplay(cm, display_labels=['Pass', 'Fail'])
    disp.plot(ax=ax, colorbar=False, cmap='Blues')
    ax.set_title(f'{name}')

plt.suptitle('Confusion Matrices', fontsize=14)
plt.tight_layout()
plt.show()ß
```

```python
# Summary accuracy table + 5-fold cross-validation
results = {}
classifiers = {'k-NN': knn, 'Decision Tree': dt, 'SVM': svm}

for name, clf in classifiers.items():
    test_acc = accuracy_score(y_test, models[name])
    cv_scores = cross_val_score(clf, X_pca, y_clean, cv=5, scoring='accuracy')
    results[name] = {
        'Test Accuracy': round(test_acc, 4),
        'CV Mean':  round(cv_scores.mean(), 4),
        'CV Std':   round(cv_scores.std(), 4)
    }

results_df = pd.DataFrame(results).T
print(results_df.to_string())
```

```python
# Bar chart comparison
results_df[['Test Accuracy', 'CV Mean']].plot(kind='bar', figsize=(9, 5),
                                               color=['steelblue', 'darkorange'],
                                               edgecolor='black')
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0.5, 1.0)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

```python
# kNN: sensitivity to k
k_values = range(1, 21)
k_accs = []
for k in k_values:
    knn_k = KNeighborsClassifier(n_neighbors=k)
    knn_k.fit(X_train, y_train)
    k_accs.append(accuracy_score(y_test, knn_k.predict(X_test)))

plt.figure(figsize=(9, 5))
plt.plot(k_values, k_accs, marker='o')
plt.title('k-NN: Accuracy vs. Number of Neighbours')
plt.xlabel('k')
plt.ylabel('Test Accuracy')
plt.xticks(k_values)
plt.tight_layout()
plt.show()
```
